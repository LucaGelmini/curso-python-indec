## Tipos de datos en Python
Los datos en Python pueden tener la forma de:
- Tipos primitivos
- Tipos no primitivos

Los tipos primitivos es la forma m√≠nima que puede tener un dato. Sobre estos tipos se pueden crear estructuras de datos m√°s complejas.
Los primitivos de Python son:
- __Integer__: numeros enteros
- __Float__: numeros decimales (o con punto flotante)
- __String__: cadenas de caracteres (texto, basicamente)
- __Boolean__: variable binaria (s√≠ o no, verdadero o falso, prendido o apagado)

A partir de estos tipos primitivos, la informaci√≥n se puede estructurar de formas m√°s complejas pero muy √∫tiles:
- __List__: conjunto indexado y desordenado de variables de cualquier tipo (primitivo o no)
- __Tuple__: son como las listas pero __inmutables__, es decir que una vez que creamos una tupla no podemos eliminar ni reemplazar sus elementos por otros.
- __Set__: son como las listas pero todos sus elementos son del mismo tipo, el orden de los elementos es aleatorio y no deja guardar elementos repetidos. A veces es util convertir una lista en set cuando queremos r√°pidamente eliminar los repetidos. Son m√°s eficientes para guardar info.
- __Dictionary__: los diccionarios guardan datos en la forma de claves asociadas a un valor. Eso nos permite acceder a datos a trav√©s de un nombre en vez de un √≠ndice num√©rico.
- __File__: Python base permite trabajar con archivos de nuestro sistema operativo. Obviamente es m√°s f√°cil trabajar con archivos de texto (leerlos, escribirlos, eliminar cosas) que leer archivos binarios de otros programas (como excel). Para leer binarios hay muchas librerias que nos ayudan a trabajar con ellos.
- __Objetos__: Si ninguna de estas estructuras de datos te gusta, pod√©s crear una. Hasta pod√©s decirle a un objeto que tipo de acciones es capaz de hacer. Dejmoslo para otra ocaci√≥n...

## Estructuras de control de flujo

En cualquier t√≠pico lenguaje de programaci√≥n debemos poder especificar como se va a ejecutar nuestro c√≥digo. No siempre queremos que se ejecuten todas las sentencias que escribimos. A veces vamos a querer que algunos procesos se ejecuten condicionalmente, otras veces vamos a querer repetir muchas veces el mismo proceso.

Estas estructuras tienen la forma de una sentencia con una palabra reservada (if, for, while, etc) y un cuerpo que corresponde a las lineas a ejecutar dentro de ese bloque de control. El cuerpo de estas estructuras debe estar indentado, con `tab` o `espacio` (en general solemos usar cuatro espacios), tratando de ser consistente en la indentaci√≥n.




__Lectura recomendada:__ [Sentencias de control de flujo](https://www.educative.io/answers/what-are-control-flow-statements-in-python)

### Condicionales
La palabra reservada es if, la condici√≥n debe ser una expresi√≥n equivalente a un `boolean`, es decir True o False. Si queremos que se ejecute algo en caso de no cumplirse la condici√≥n podemos escribir `else`. Si queremos que se revisen en serie varias condiciones podemos usar la palabra reservada `elif`.

__Veamos un ejemplo:__

```python
# Si A es igual a B
A="foo"
B="baz"

if A == B:
    print("son iguales")
else:
    print("No son iguales")
```

```python
C="foo"

if A == B:
    print("A es igual a B")
elif A == C:
    print("A es igual a C")
else:
    print("no hay nada igual a A")
```

```python
id(A) == id(B)
```

NOTA: para estas comparaciones podriamos usar la palabra reservada `is` en lugar de `==` pero es recomendable usar `==` si quieren asegurarse que lo que se est√° comparando es el valor de esas variables. `is` compara si es la misma variable; o, dicho de modo brusco, compara si es el mismo cacho de memoria. En este caso, como Python no es tonto, no va a gastar memoria para guardar exactamente lo mismo en A y B.

En general va a funcionar perfectamente, pero cuando esten usando estructuras de datos mas complejas puede que se comporte distinto a lo que esperan.

### Bucles for y while

A veces tenemos que repetir el mismo proceso m√∫ltiples veces. Para eso vamos usar estos bucles.

#### Bucle for
En Python tenemos estas cosas a las que le llamamos iterables, b√°sicamente son tipos complejos que pueden tener un √≠ndice que permita acceder a muchas variables distintas. Hay unos iterables m√°s raros que no vamos a tocar en este curso. ¬°Pero hay otros iterables que ya vimos!

Las listas, tuplas, sets y diccionarios SON iterables. Si queremos hacer una operaci√≥n por cada elemento de estos, vamos a usar un bucle for. Veamos un ejemplo de lista y otro de diccionario:

```python
lista_de_enteros = [10,20,30,40,50]

for numero in lista_de_enteros:
    print(numero + 1)
```

```python
# ¬øY como hacemos para modificar la lista?

# Paso 1:
# Podemos acceder a los elementos de la lista por su √≠ndice (EMPIEZA POR 0 COMO EL DIOS DE LA PROGRAMACI√ìN MANDA, no como los degenerados de R que arrancan por 1...)

lista_de_enteros[3]
```

```python
# Ah pero cuando recorra la lista tengo que modificar cada elemento, pero para ello tengo que saber su √≠ndice al momento de recorrerla
# Paso 2: enumerate al rescate!
# enumarate es una funci√≥n de Python base que toma un 
# iterable y se convierte en otro iterable solo que ahora sus elementos son tuplas compuestas por el elemento original m√°s su √≠ndice
for (indice,  numero) in enumerate(lista_de_enteros):
    print(f"√çndice: {indice}, Numero originial: {numero}")
```

```python
# Paso 3: asignar los nuevos numeros a la lista original

for (indice,  numero) in enumerate(lista_de_enteros):
    lista_de_enteros[indice] = numero + 1
    
print(lista_de_enteros)
```

```python
# Si quieren poner esos valores en una nueva lista pueden crear una nueva y usar el m√©todo append (como lo sugiere la palabra en ingl√©s, solo le agrega nuevas cosas al final)
nueva_lista = [] # Es una lista vacia!
for numero in lista_de_enteros:
    nueva_lista.append(numero - 5)
    
nueva_lista
```

```python
# Pro tip: Hay una forma que al principio es media confuza para hacer esto escribiendo menos, se las dejo para que sepan que existe pero no le presten atenci√≥n
# en ingl√©s se llama comprehension. En este caso es list comprehension.

lista_de_enteros = [numero/10 for numero in lista_de_enteros]

print(lista_de_enteros)
```

### Bucle while
A veces no tenemos un iterable como los anteriores pero a√∫n as√≠ queremos repetir un proceso hasta que determinada condici√≥n sea True. Para eso usamos el bucle while

```python
# Contemos hasta 100 con while
a = 0
while a != 100:
    a+=1
    
print(a)
```

¬øQu√© pasa si queremos escapar de los bucles por m√°s que no se cumpla la condici√≥n? Usamos la palabra reservada `break`.

```python
while True: # Este bucle es infinito
    a-=1 # vamos a ir restandole 1 al valor de a que ahora quedo en 100
    if a == 11: # cuando a sea igual a 11, salimos del bucle
        break
    
print(a)
```

### ¬°Funciones!

En el mundo de la programaci√≥n tratamos de seguir una regla a la que llamamos DRY (don't repeat yourself), es decir, "no te repitas". En este mundo, la regla del menor esfuerzo es nuestra mayor virtud, cuanto menos c√≥digo escribamos para resolver una tarea, menos probabilidades de introducir errores vamos a generar.

Para eso usamos las funciones, encapsulamos fragmentos de c√≥digo reutilizable para solo escribirlo una vez y controlar mejor las condiciones en las que se ejecuta. Otro consejo que les doy, es que cuando empiecen a escribir un programa, controlen la cantidad de variables que usan, y las variables de las que depende cada parte de su proceso. No queremos llevarnos la sorpresa de que una variable escrita 500 l√≠neas atr√°s genere un problema en un proceso que no tiene nada que ver. En eso tambi√©n las funciones nos ayudan much√≠simo.

Ya estuvimos usando una funci√≥n durante toda esta clase: ¬°`print()`!. En este caso, las cosas que queremos imprimir por consola se las damos a la funci√≥n `print` como _argumentos_ y `print`, como si fuera una caja negra de la cu√°l desconocemos su funcionamiento, imprime esos argumentos en la consola.

#### Usar funciones ya definidas
Cuando usamos una funci√≥n sin par√©ntesis tenemos que pensarla como una variable m√°s: una variable del tipo `Callable` (es el tipo de dato especifico de las funciones). Pero cuando le ponemos los par√©ntesis, como en print, es cuando sucede la magia y la funci√≥n propiamente "funciona"; o sea que hace las operaciones que tiene definidas.
- Hay funciones que no piden ning√∫n argumento, entonces para llamarlas solo hay que ponerle los parentesis vac√≠os: `funcion_sin_args()`
- Hay funciones como `print(arg1, arg2, arg3, ...)` que podemos pasarle uno o muchos argumentos para que imprima por consola
- Y hay funciones que adem√°s pueden devolver alg√∫n valor. Imaginemos una funci√≥n que suma dos numeros: `sumar(num1, num2)`. La podemos usar de la siguiente forma `resultado = suma(1,2)`. En ese caso `resultado` ser√° igual a 4.

#### Definir funciones propias
¬°Lo m√°s interesante es que podemos definir funciones que ejecuten el c√≥digo que nosotros querramos!
Para definirlas debemos empezar con la palabra reservada `def`, luego el nombre y los par√©ntesis. Pero en este caso a lo que est√° dentro de los par√©ntesis lo llamamos _par√°metros_. Veamos ejemplos:


```python
# Definamos una funci√≥n que imprima en consola un mensaje pero que antes nos salude
def saludos_y_mensaje(mensaje): # Esta funci√≥n tiene un solo par√°metro que debera ser in string
    print('Hola '+mensaje)
    
# Usemos esta funci√≥n con un mensaje cualquiera
saludos_y_mensaje("Mart√≠n") # Le pasamos "Mart√≠n" como argumento
```

```python
# Definamos otra funci√≥n que haga alguna operaci√≥n aritm√©tica y que nos devuelva el resultado
# Hagamos que haga el promedio de todos los n√∫meros de una lista

def promedio_de_lista(lista):
    # Tenemos que preguntar por la cantidad de elementos
    cantidad = len(lista) # Python base tiene esta funci√≥n que devuelve un n√∫mero entero con el largo de un iterable
    suma = sum(lista)
    promedio = suma/cantidad
    return promedio # return es la palabra reservada que indica cual es el valor que la funci√≥n debe devolver


prom = promedio_de_lista([10,20])
print(prom)
```

Como ven, al igual que las estructuras de control de flujo que ya vimos (if, while, for, etc.), el cuerpo  de la funci√≥n tambi√©n est√° indentado. En Python la indentaci√≥n indica un bloque de c√≥digo, en el caso de las funciones eso implica que los par√°metros y cualquier otra variable que definamos adentro de ellas tendr√°n un alcance local; es decir que solo tienen efecto dentro de la funci√≥n y no son accesibles por fuera. 

```python
def una_funcion(foo):
    bar = 'leviosa'
    print(foo+' '+bar)
    
una_funcion('wingardium')
```

```python
# Pero la variable bar no existe fuera de la funci√≥n
print(bar)
```

Esto no es as√≠ en el caso contrario: si una variable est√° definida fuera de una funci√≥n, esta puede ser usada dentro de ella.
