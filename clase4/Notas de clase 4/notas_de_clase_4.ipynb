Supongamos que quiero calcular la evolución del precio relativo de algunas categorías frente al nivel general.

```python
# Si lo quisieramos repetir para la columna que querramos:
def calcula_ratio_categoria(columna:str) -> pd.Series: 
    '''Ingresamos alguna de las columnas de la dataframe
    para encontrar su ratio'''
    ratio_calculado = ipc_df[columna] / ipc_df["Nivel general"] #Dividí una columna por la otra
    return ratio_calculado #No se lo olviden!!!
calcula_ratio_categoria(columna="Prendas de vestir y calzado")
```

Para no repetirnos. Qué tal si hacemos una función?

Para ello. ¿Qué es lo único que cambia? La ruta del archivo y el nombre de la columna de la serie. 
Momentáneamente la nombro como "Serie" para explicar lo que es el merge. 

```python
def limpiar_series_dest_tendencia(ruta: str):
    serie = (pd.read_excel(ruta,
                                header=2)
                    .dropna(axis=0, thresh=2)
                    .drop("Totales", axis = 1)
                    .replace("…", np.nan))

    serie = serie.melt(id_vars=['Año'], 
                        value_vars=serie.columns[1:], #Todas las columnas menos "Año" 
                        var_name='Mes', 
                        value_name='Serie')
    return serie
```

```python
serie_desest = limpiar_series_dest_tendencia(ruta="./data/expo_desestacionalizada_1996_2023.xls")
serie_tendencia = limpiar_series_dest_tendencia(ruta="./data/expo_tendencia_ciclo_1996_2023.xlsx")
```

Las DataFrame ya están listas para ser procesadas en base a lo que necesitemos.
Si queremos hacer un "join" (similar al buscarv en excel):



```python
serie_mergeada = serie_desest.merge(right=serie_tendencia, on=["Año","Mes"], how="left", suffixes=[" desestacionalizada", " tendencia-ciclo"])
```

```python
#Leo y filtro las columnas que me interesan.
presidencias_argentinas = pd.read_html("https://es.wikipedia.org/wiki/Anexo:Presidentes_de_la_Naci%C3%B3n_Argentina")
presidencias_argentinas = presidencias_argentinas[0] #Lee varias tablas. La primera es la que interesa.
presidencias_argentinas = presidencias_argentinas.iloc[:-1, 2:-3] #El último presidente todavía no asumió
presidencias_argentinas = presidencias_argentinas.rename({"Presidente de la Nación.2" : "Presidente"}) #Renombro la columna

#Elimino las filas que están repetidas a través de las columnas
def check_duplicate_row(row):
    '''Me quedo con las filas cuyos valores se repiten en esa misma fila,
    pero para todas las columas'''
    return all(x == row[0] for x in row)
filas_duplicadas = presidencias_argentinas.apply(check_duplicate_row, axis=1)
presidencias_argentinas = presidencias_argentinas[~filas_duplicadas] 

#Elimino duplicados y reseteo el índice
presidencias_argentinas = (presidencias_argentinas.drop_duplicates() #Eliminé los duplicados
                           .reset_index(drop = True)) #Resetié el índice

#Cambio de formato a datetime en las columnas de inicio y fin del gobierno. 
meses_ingles = {'enero': 'January', 'febrero': 'February', 'marzo': 'March', 'abril': 'April', 'mayo': 'May', 'junio': 'June', 
         'julio': 'July', 'agosto': 'August', 'septiembre': 'September', 'octubre': 'October', 'noviembre': 'November', 'diciembre': 'December'}
presidencias_argentinas['Inicio del gobierno'] = (presidencias_argentinas['Inicio del gobierno'].str.replace(r'\[.*\]', '', regex=True) #Elimino lo que está en []
                                                  .str.replace('\u2009\u200b', '') #Habían caracteres invisibles
                                                  .apply(lambda x: ' '.join([meses_ingles[word] if word in meses_ingles else word for word in x.split()]))) #Cambio el nombre de los meses a inglés

presidencias_argentinas['Inicio del gobierno'] = pd.to_datetime(presidencias_argentinas['Inicio del gobierno'], format='%d de %B de %Y', errors='coerce')

presidencias_argentinas['Fin del gobierno'] = (presidencias_argentinas['Fin del gobierno'].str.replace(r'\[.*\]', '', regex=True) #Elimino lo que está en []
                                                  .str.replace('\u2009\u200b', '')
                                                  .apply(lambda x: ' '.join([meses_ingles[word] if word in meses_ingles else word for word in x.split()])))
presidencias_argentinas['Fin del gobierno'] = pd.to_datetime(presidencias_argentinas['Fin del gobierno'], format='%d de %B de %Y', errors='coerce')

#Agrego la fecha de terminación del mandato de Alberto Fernández. 
presidencias_argentinas.iloc[57,4] = pd.to_datetime("2023-12-10", format="%Y-%m-%d")

#Grafico
fig = go.Figure()
fig.add_trace(go.Histogram(x = presidencias_argentinas["Lugar en el que nació"]))
fig.update_xaxes(categoryorder='total descending')
fig.update_layout(template = None, title_text = "Presidencias según lugar de nacimiento")
```
